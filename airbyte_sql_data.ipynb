#####QUERY DE CONSULTAS DE TABELAS INGERIDAS PELO AIRBYTE

```python
%sql

SELECT
    _airbyte_data:CustomerID AS CustomerID,
    _airbyte_data:FirstName AS FirstName,
    _airbyte_data:LastName AS LastName,
    _airbyte_data:Email AS Email,
    _airbyte_data:Ocuppation AS Occupation,
    _airbyte_data:City AS City,
    _airbyte_data:Country AS Country,
    _airbyte_data:CreatedAt AS CreatedAt
FROM lakehouse.airbyte_internal.default_raw__stream_customers;
```

```python
%sql


select * from lakehouse.airbyte_internal.default_raw__stream_orderdetails;

SELECT
  _airbyte_data:OrderDetailID::int AS OrderDetailID,
  _airbyte_data:OrderID::int AS OrderID,
  _airbyte_data:ProductID::int AS ProductID,
  _airbyte_data:Quantity::int AS Quantity,
  _airbyte_data:UnitPrice::decimal(10,2) AS UnitPrice
FROM lakehouse.airbyte_internal.default_raw__stream_orderdetails;
```

```python
%sql
SELECT
  _airbyte_data:OrderID AS OrderID,
  _airbyte_data:CustomerID AS CustomerID,
  _airbyte_data:Total AS Total,
  _airbyte_data:OrderDate AS OrderDate
FROM lakehouse.airbyte_internal.default_raw__stream_orders;
```

```python
%sql


SELECT
  _airbyte_data:ProductID::int AS ProductID,
  _airbyte_data:ProductName::string AS ProductName,
  _airbyte_data:Category::string AS Category,
  _airbyte_data:UnitPrice::decimal(10,2) AS UnitPrice,
  _airbyte_data:Weight::decimal(10,2) AS Weight,
  _airbyte_data:CreatedAt::timestamp AS CreatedAt
FROM lakehouse.airbyte_internal.default_raw__stream_products;
```
